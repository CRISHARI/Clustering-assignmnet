```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i,n_init=20)
  km.fit(X)
  css=km.inertia_
  print('clusters=',i,"wcss=",css)
  wcss.append(css)
```

##DBSCAN  Clustering and Implementation

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

# Generating sample data (you would replace this with your actual data)
X, _ = make_blobs(n_samples=178, centers=5, cluster_std=0.60, random_state=42)

# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X)

# Plotting the results
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

for k, col in zip(unique_labels, colors):
    if k == -1:  # Noise points are labeled as -1
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask]
    plt.scatter(xy[:, 0], xy[:, 1], c=[col], edgecolors='k', s=50)

plt.title('DBSCAN Clustering')
plt.show()
```
